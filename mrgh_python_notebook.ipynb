```python
import numpy as np
```

```python
points = np.arange(-5,5, 0.01)
```

import matplotlib.pyplot as plt

```python
xs, ys = np.meshgrid(points, points)
```

```python
z = np.sqrt(xs ** 2 + ys ** 2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(z, cmap=plt.cm.gray);plt.colorbar()
```
