```python
adjusted_predictors = [pred for pred in predictors if pred != 'bloodserum03']
adjusted_predictors
```

Preparing to make a baseline regressor with these predictors.

```python
x = df[adjusted_predictors].values
y = df.y.values

reg = LinearRegression()
reg.fit(x, y)
ypred = reg.predict(x)
rmse = np.sqrt(mean_squared_error(ypred, y))
print("RMSE: %0.2f" % (rmse.mean()))
```

   KFold Analysis With Cross Validation

```python
cv_errors = []
kfold = KFold(len(df), n_folds=10)
nfeatures = range(1, len(adjusted_predictors))

for nfeature in nfeatures:
    # build model with varying number of features
    selector = SelectKBest(f_regression, k=nfeature)
    selector.fit(x, y)
    selected = selector.get_support()
    feats = [col for (col,sel) in zip(predictors, selected) if sel]
    x_r = df[feats].values
    y = df["y"].values
    rmses = []
    for train, test in kfold:
        # each model is cross validated 9 times
        xtrain, ytrain,xtest, ytest = x_r[train], y[train], x_r[test], y[test]   
        reg = LinearRegression()
        reg.fit(xtrain, ytrain)
        ypred = reg.predict(xtest)
        rmses.append(np.sqrt(mean_squared_error(ypred, ytest)))
    cv_errors.append(np.mean(rmses))
    
plt.plot(nfeatures, cv_errors)
plt.xlabel("number of features")
plt.ylabel("RMSE")
```
